```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
```

```python
# funÃ§Ãµes do tipo z = f(x,y), visualizaÃ§Ã£o das curvas e retas tangentes
# z = f(x,y) = 4 - xÂ² - 2yÂ² 
# Determinar fx(1,1) e fy(1,1)  P = (1,1) neste ponto z = 1
eixox = np.arange(-2, 2, 0.1)   # x variando de -2 a 2
eixoy = np.arange(-np.sqrt(2), np.sqrt(2), 0.1)   # y variando de -raiz(2) a raiz(2)
x, y = np.meshgrid(eixox,eixoy)  # cria a grade (domÃ­nio)
z = 4 - x**2 - 2*(y**2)  # z Ã© a funÃ§Ã£o: parabolÃ³ide

# o plano y = 1 intercepta o parabolÃ³ide 
eixoz = np.arange(-4, 3, 0.1)
xplanoy, zplanoy = np.meshgrid(eixox,eixoz)
# resultando na curva na C1: z = 2 - xÂ²
C1 = 2 - x**2
# T1 Ã© a reta tangente Ã  curva C1 em P
# a inclinaÃ§Ã£o de T1 Ã© fx = -2x e T1 passa por P = (1,1,1)
T1 = -2*x + 3

# o plano x = 1 intercepta o parabolÃ³ide 
eixoz = np.arange(-4, 3, 0.1)
yplanox, zplanox = np.meshgrid(eixoy,eixoz)
# resultando na curva C2: z = 3 - 2yÂ²
C2 = 3 - 2*(y**2)
# T2 Ã© a reta tangente Ã  curva C2 em P
# a inclinaÃ§Ã£o de T2 Ã© fy = -4x e T2 passa por P = (1,1,1)
T2 = -4*y + 5

# Plano tangente ao parabolÃ³ide em P = (1,1,1)
plano = -2*x - 4*y + 7

fig = plt.figure(figsize=(6,6))          # cria o objeto figura
ax = Axes3D(fig)            # cria eixos 3d para a figura
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  
ax.plot_wireframe(x, y, z, rstride=1, cstride=1, linewidth=1, color='b') 
ax.plot_surface(xplanoy, 1, zplanoy, rstride=1, cstride=1, cmap=cm.viridis)
ax.plot_wireframe(x, 1, C1, rstride=1, cstride=1, linewidth=5, color='k') 
ax.plot_surface(1, yplanox, zplanox, rstride=1, cstride=1, cmap=cm.viridis)
ax.plot_wireframe(1, y, C2, rstride=1, cstride=1, linewidth=5, color='k')  
ax.plot_wireframe(x, 1, T1, rstride=1, cstride=1, linewidth=5, color='r') 
ax.plot_wireframe(1, y, T2, rstride=1, cstride=1, linewidth=5, color='r') 
ax.plot_wireframe(x, y, plano, rstride=2, cstride=2, linewidth=1, color='g')  

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')
plt.show()
```
