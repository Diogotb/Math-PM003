```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 

# variaÃ§Ã£o do parÃ¢metro t
t = np.linspace(0, 2*np.pi, 100)
# equaÃ§Ãµes paramÃ©tricas
xt = 10*np.cos(t)
yt = 10*np.sin(t)

fig, ax = plt.subplots()  # cria objeto figura e associa eixos
plt.plot(xt,yt)           # faz o grÃ¡fico
ax.set_aspect('equal')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
plt.show()
```

```python
xt = 10*np.cos(t)
yt = 10*np.sin(t)

xd, yd = np.meshgrid(xt, yt)     # cria a grade (produto cartesiano), todo o domÃ­nio da funÃ§Ã£o h(x,y)
hxy = 4 + 0.01*(xd**2 - yd**2)     # define h(x,y)

fig1 = plt.figure(figsize=(4,4))           # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(xd, yd, hxy, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
plt.show()
```

```python
# variaÃ§Ã£o do parÃ¢metro t
t = np.linspace(0, 2*np.pi, 100)

# equaÃ§Ãµes paramÃ©tricas
xt = 10*np.cos(t)
yt = 10*np.sin(t)
ht = 4 + 0.01*(xt**2 - yt**2)

# em todo domÃ­nio
xd, yd = np.meshgrid(xt, yt)  # cria a grade 
hxyd = 4 + 0.01*(xd**2 - yd**2)

fig2 = plt.figure(figsize=(4,4))           # cria o objeto figura
ax2 = Axes3D(fig2)                         # cria eixos 3d para a figura

# funÃ§Ã£o h(x,y)
ax2.plot_wireframe(xd, yd, hxyd, rstride=3, cstride=3, linewidth=0.5, color='b')

# curva no R3: (xt, yt, 0)
ax2.plot(xt, yt, 0, 'g')

# curva no R3: (xt, yt, ht)
ax2.plot(xt, yt, ht, 'r', linewidth=2)

ax2.set_xlabel('eixo x')
ax2.set_ylabel('eixo y')
ax2.set_zlabel('eixo z')
plt.show()
```

```python
# variaÃ§Ã£o do parÃ¢metro t
t = np.linspace(0, 2*np.pi, 100)

# equaÃ§Ãµes paramÃ©tricas
xt = 10*np.cos(t)
yt = 10*np.sin(t)
ht = 4 + 0.01*(xt**2 - yt**2)

fig3 = plt.figure(figsize=(4,4))           # cria o objeto figura
ax3 = Axes3D(fig3)                         # cria eixos 3d para a figura

# curva no R3: (xt, yt, 0)
ax3.plot(xt, yt, 0, 'g', linewidth=5)

# curva no R3: (xt, yt, ht)
ax3.plot(xt, yt, ht, 'r', linewidth=5)

# parametrizando pontos de r0 atÃ© r1 para plotar a "cortina"
alpha = np.linspace(0, 1, 30)
for i in range(len(xt)):
    r0 = np.array([xt[i], yt[i], 0])
    r1 = np.array([xt[i], yt[i], ht[i]])
    for a in alpha:
        r = (1 - a)*r0 + a*r1
        plt.plot([r[0]], [r[1]], [r[2]], 'b.', markersize=2)

ax3.set_xlabel('eixo x')
ax3.set_ylabel('eixo y')
ax3.set_zlabel('eixo z')
plt.show()
```
