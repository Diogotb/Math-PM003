```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
from math import e
import math
```

```python
#exerccio 23- plao tangente ao cone z=x+y lista 1 plotando grficos x, y, z
#grficos de z = f(x,y)
eixox = np.arange(-10, 10.5, 0.5)  # min, max, passo
eixoy = np.arange(-10, 10.5, 0.5)  # min, max, passo

x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domnio)
z = (x**2 + y**2)**1/2


fig1 = plt.figure(figsize=(4,4))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
plt.show()

# curvas de nvel
fig2, ax2 = plt.subplots(figsize=(4,4))
ax2.contour(x,y,z,50) # 50 curvas de nvel
plt.show()
```
