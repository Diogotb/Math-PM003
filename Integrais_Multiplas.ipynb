```python
# Exercício 14.4.6
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

# variação do parâmetro t
t = np.linspace(np.pi/2, np.pi, 100)
# equações paramétricas
r = 2*np.sin(t)
xt = r*np.cos(t)
yt = r*np.sin(t)

fig, ax = plt.subplots()  # cria objeto figura e associa eixos
plt.plot(xt, yt, 'r', linewidth=3)           # faz o gráfico

pontosx = np.linspace(-2, 2, 100)
pontosy = np.linspace(-2, 2, 100)
eixox = [0]*len(pontosy)
eixoy = [0]*len(pontosx)
plt.plot(pontosx, eixoy, 'k', linewidth=3) 
plt.plot(eixox, pontosy, 'k', linewidth=3) 

ax.set_aspect('equal')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
```

```python
# Exercício 14.4.18

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

# variação do parâmetro t
t = np.linspace(0, 2*np.pi, 100)

# equações paramétricas
r1 = 3*np.cos(t)
xt1 = r1*np.cos(t)
yt1 = r1*np.sin(t)

# equações paramétricas
r2 = 1 + np.cos(t)
xt2 = r2*np.cos(t)
yt2 = r2*np.sin(t)

fig, ax = plt.subplots()  # cria objeto figura e associa eixos
plt.plot(xt1, yt1, 'r', linewidth=3)           # faz o gráfico
plt.plot(xt2, yt2, 'r', linewidth=3)           # faz o gráfico

pontosx = np.linspace(-3, 3, 100)
pontosy = np.linspace(-3, 3, 100)
eixox = [0]*len(pontosy)
eixoy = [0]*len(pontosx)
plt.plot(pontosx, eixoy, 'k', linewidth=3) 
plt.plot(eixox, pontosy, 'k', linewidth=3) 

ax.set_aspect('equal')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.show()
```

```python
# Exercício 15.2.23
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

# em todo domínio
xd, yd = np.meshgrid(x, y)  # cria a grade 
z = 4 - xd - 2*yd

fig2 = plt.figure(figsize=(4,4))           # cria o objeto figura
ax2 = Axes3D(fig2)                         # cria eixos 3d para a figura

ax2.plot_wireframe(xd, yd, z, rstride=3, cstride=3, linewidth=0.5, color='b')

ax2.set_xlabel('eixo x')
ax2.set_ylabel('eixo y')
ax2.set_zlabel('eixo z')
plt.show()
```

```python
# Exercício 15.1.14
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(0, 4, 100)
y = np.linspace(0, 2, 100)

# em todo domínio
xd, yd = np.meshgrid(x, y)  # cria a grade 
z = np.sqrt(9 - yd**2)

fig2 = plt.figure(figsize=(4,4))           # cria o objeto figura
ax2 = Axes3D(fig2)                         # cria eixos 3d para a figura

ax2.plot_wireframe(xd, yd, z, rstride=3, cstride=3, linewidth=0.5, color='b')

ax2.set_xlabel('eixo x')
ax2.set_ylabel('eixo y')
ax2.set_zlabel('eixo z')
plt.show()
```
