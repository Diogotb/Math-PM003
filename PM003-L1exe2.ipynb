```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
```

```python
#exerccio 2-e lista 1 plotando grficos x, y, z
# grficos de z = f(x,y) com descontinuidades
# eixox = np.arange(-10, 10, 0.5)  # no! cuidado, descontnua em (0,0)
# eixoy = np.arange(-10, 10, 0.5)  # no! cuidado, descontnua em (0,0)
x1 = np.arange(-10, -0.5, 0.5) # -10 at -0.5 
x2 = np.arange(1.5, 10, 0.5)   # 0.5 at 10
eixox = np.append(x1, x2)
y1 = np.arange(-10, 1.5, 0.5)   # -10 at -0.5  
y2 = np.arange(2.5, 10, 0.5)     # 0.5 at 10
eixoy = np.append(y1, y2)
x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domnio)
z = (x*y-y)/((x-1)**2+y**2)    # z = f(x,y)

fig1 = plt.figure(figsize=(4,4))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
plt.show()

# curvas de nvel
fig2, ax2 = plt.subplots(figsize=(4,4))
ax2.contour(x,y,z,50) # 50 curvas de nvel
ax2.plot([0],[0], "ro")
plt.show()
```
