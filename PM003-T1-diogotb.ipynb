```python
def sistema (variaveis):  # definição do sistema
    (x, y, L) = variaveis
    eq1 = 2*(x+2) - L*2*(x+1)
    eq2 = 6*(y-3) - L*2*(y-1)
    eq3 = (x+1)**2+(y-1)**2 - 9
    return [eq1, eq2, eq3]

s0 = np.array([-1,-2,0])  # solução inicial 1
s = fsolve(sistema, s0)
print("P10 = ", s)

s0 = np.array([-3,3,0])   # solução inicial 2
s = fsolve(sistema, s0)
print("P11 = ", s)
```
