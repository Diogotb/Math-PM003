```python
import numpy as np
import math
from math import pi
from math import e
from math import sqrt
from sympy import *
x, y, z, r = symbols('x y z r')
dx, dy, dz = symbols('dx dy dz')
t, phi, k = symbols('t phi k')
u, v = symbols('u v')
init_printing(use_unicode=True)
```

```python
var('x y z r')
var('t phi')
f = input('f: ')
f=str(simplify(f))
print(f)
for i in range(3):
    esp = int(input('Resolver por métodos especiais: 1-Sim, 2-Não'))
    if esp==1:
        cp = int(input('1-Cord.Polares;\n2-Cord.Cilindricas;\n3-Cord.Esfericas;\n4-Linha;\n5-Fd;\n6-Green;\n7-Parametrização;\n8-Area Superfície;\n9-Stokes;\n10-Divergente:\n'))
    else:
        grau = int(input('grau de integracao: '))
        break
    if cp==1:
        x=r*cos(t)
        y=r*sin(t)
        f=eval(f)*r
        print(f)
        f=simplify(f)
        print(f)
        grau=2
        break
    if cp==2:
        x=r*cos(t)
        y=r*sin(t)
        f=eval(f)*r
        print(f)
        f=simplify(f)
        print(f)
        grau=3
        break
    if cp==3:
        x=r*cos(t)*sin(phi)
        y=r*sin(t)*sin(phi)
        z=r*cos(phi)
        f=eval(f)*sin(phi)*r**2
        print(f)
        f=simplify(f)
        print(f)
        grau=3
        break
    if cp==4:
        x=input('xt: ')
        x = simplify(x)
        y=input('yt: ')
        y = simplify(y)
        z=input('zt: ')
        z = simplify(z)
        fl=sqrt(diff(x)**2 +diff(y)**2 +diff(z)**2)
        print(fl)
        f= eval(f)*fl
        f=simplify(f)
        print(f)
        grau=1
        break
    if cp==5:
        x=input('xt: ')
        x = simplify(x)
        dx = diff(x)
        y=input('yt: ')
        y = simplify(y)
        dy = diff(y)
        z=input('zt: ')
        z = simplify(z)
        dz = diff(z)
        f= eval(f)
        f=str(simplify(f))
        print(f)
    if cp==6:
        P=input('P(x,y): ')
        Q=input('Q(x,y): ')
        f = str(diff(Q,x)-diff(P,y))
        print(f)
    if cp==7:
        x=input('xt: ')
        x = simplify(x)
        y=input('yt: ')
        y = simplify(y)
        z=input('zt: ')
        z = simplify(z)
        f= eval(f)
        print(f)
        f=str(simplify(f))
        print(f)
    if cp==8:
        print(diff(f,x))
        print(diff(f,y))
        f = str(sqrt(1+diff(f,x)**2+diff(f,y)**2))
        print(f)
    if cp==9:
        P=input('P(x,y,z): ')
        Q=input('Q(x,y,z): ')
        R=input('R(x,y,z): ')
        print(diff(R,y)-diff(Q,z))
        print(diff(P,z)-diff(R,x))
        print(diff(Q,x)-diff(P,y))
        f = str((diff(R,y)-diff(Q,z))+(diff(P,z)-diff(R,x))+(diff(Q,x)-diff(P,y)))
        print(f)
    if cp==10:
        P=input('P(x,y,z): ')
        Q=input('Q(x,y,z): ')
        R=input('R(x,y,z): ')
        print(diff(P,x))
        print(diff(Q,y))
        print(diff(R,z))
        f = str((diff(P,x)+diff(Q,y))+diff(R,z))
        print(f)

g=1
while g <= grau:
    o = input('ordem: ')
    
    if o == 'x':
        inif = integrate(f,x)
        print('integral em x: ',inif)
        m1 = input('Digite o intervalo da integração em x1: ')
        inif1 = inif.subs(x,m1)
        print('subs x1: ',inif1)
        m2 = input('Digite o intervalo da integração em x2: ')
        inif2 = inif.subs(x,m2)
        print('subs x2: ',inif2)
        inif = inif2-inif1
    elif o == 'y':
        inif = integrate(f,y)
        print('integral em y: ',inif)
        m1 = input('Digite o intervalo da integração em y1: ')
        inif1 = inif.subs(y,m1)
        print('subs y1: ',inif1)
        m2 = input('Digite o intervalo da integração em y2: ')
        inif2 = inif.subs(y,m2)
        print('subs y2: ',inif2)
        inif = inif2-inif1
    elif o == 'z':
        inif = integrate(f,z)
        print('integral em z: ',inif)
        m1 = input('Digite o intervalo da integração em z1: ')
        inif1 = inif.subs(z,m1)
        print('subs z1: ',inif1)
        m2 = input('Digite o intervalo da integração em z2: ')
        inif2 = inif.subs(z,m2)
        print('subs z2: ',inif2)
        inif = inif2-inif1
    elif o == 'r':
        inif = integrate(f,r)
        print('integral em r: ',inif)
        m1 = input('Digite o intervalo da integração em r1: ')
        inif1 = inif.subs(r,m1)
        print('subs r1: ',inif1)
        m2 = input('Digite o intervalo da integração em r2: ')
        inif2 = inif.subs(r,m2)
        print('subs r2: ',inif2)
        inif = inif2-inif1
    elif o == 't':
        inif = integrate(f,t)
        print('integral em t: ',inif)
        m1 = input('Digite o intervalo da integração em t1: ')
        inif1 = inif.subs(t,m1)
        print('subs t1: ',inif1)
        m2 = input('Digite o intervalo da integração em t2: ')
        inif2 = inif.subs(t,m2)
        print('subs t2: ',inif2)
        inif = inif2-inif1
    elif o == 'u':
        inif = integrate(f,u)
        print('integral em u: ',inif)
        m1 = input('Digite o intervalo da integração em u1: ')
        inif1 = inif.subs(u,m1)
        print('subs u1: ',inif1)
        m2 = input('Digite o intervalo da integração em u2: ')
        inif2 = inif.subs(u,m2)
        print('subs u2: ',inif2)
        inif = inif2-inif1
    elif o == 'v':
        inif = integrate(f,v)
        print('integral em v: ',inif)
        m1 = input('Digite o intervalo da integração em v1: ')
        inif1 = inif.subs(v,m1)
        print('subs v1: ',inif1)
        m2 = input('Digite o intervalo da integração em v2: ')
        inif2 = inif.subs(v,m2)
        print('subs v2: ',inif2)
        inif = inif2-inif1
    else:
        inif = integrate(f,phi)
        print('integral em phi: ',inif)
        m1 = input('Digite o intervalo da integração em phi1: ')
        inif1 = inif.subs(phi,m1)
        print('subs phi1: ',inif1)
        m2 = input('Digite o intervalo da integração em phi2: ')
        inif2 = inif.subs(phi,m2)
        print('subs phi2: ',inif2)
        inif = inif2-inif1
    print('valor: ',inif)
    inif = simplify(inif)
    print('valor: ',inif)
    f = inif
    g +=1
```
