```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
import math
from math import e
from sympy import *
x, y, z = symbols('x y z')
a, b, r= symbols('a b r')
theta = symbols('theta')
init_printing(use_unicode=True)
```

```python
#exercÃ­cio 4-a lista 1 plotando grÃ¡ficos x, y, z
# grÃ¡ficos de z = f(x,y) com descontinuidades
# eixox = np.arange(-10, 10, 0.5)  # nÃ£o! cuidado, descontÃ­nua em (0,0)
# eixoy = np.arange(-10, 10, 0.5)  # nÃ£o! cuidado, descontÃ­nua em (0,0)
x1 = np.arange(-10, -0.5, 0.5) # -10 atÃ© -0.5 
x2 = np.arange(1.5, 10, 0.5)   # 0.5 atÃ© 10
eixox = np.append(x1, x2)
y1 = np.arange(-10, 1.5, 0.5)   # -10 atÃ© -0.5  
y2 = np.arange(2.5, 10, 0.5)     # 0.5 atÃ© 10
eixoy = np.append(y1, y2)
eixoz = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domÃ­nio)
z = ((x**2)*(y**3))/(2*x**2 +y**2)   # z = f(x,y)

fig1 = plt.figure(figsize=(6,6))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
plt.show()

# curvas de nÃ­vel
fig2, ax2 = plt.subplots(figsize=(4,4))
ax2.contour(x,y,z,50) # 50 curvas de nÃ­vel
ax2.plot([0],[0], "ro")
plt.show()
```

```python
#cÃ¡lculo do limite de f(x,y) por coordenadas polares 
z = ((x**2)*(y**3))/(2*x**2 +y**2)
x = r*cos(theta)
y = r*sin(theta)
limit(z,r,0)
```
