```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
from math import e
import math
```

```python
#exercício 1- b lista 1 plotando gráficos x, y, z
#gráficos de z = f(x,y)
x1 = np.arange(-10, -0.5, 0.5) # -10 até -0.5 
x2 = np.arange(1.5, 10, 0.5)   # 0.5 até 10
eixox = np.append(x1, x2)
y1 = np.arange(-10, 1.5, 0.5)   # -10 até -0.5  
y2 = np.arange(2.5, 10, 0.5)     # 0.5 até 10
eixoy = np.append(y1, y2)
x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domínio)
#if x==0 and y==0:
#z = 0   # z = f(x,y)
#else:
z = ((x**3)*y-x*(y**3))/((x**2)+(y**2))   # z = f(x,y)

fig1 = plt.figure(figsize=(4,4))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
plt.show()

# curvas de nível
fig2, ax2 = plt.subplots(figsize=(4,4))
ax2.contour(x,y,z,50) # 50 curvas de nível
plt.show()
```
