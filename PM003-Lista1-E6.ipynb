```python
#exercício 6 lista 1 plotando gráficos x, y, z
# funções do tipo z = f(x,y), visualização das curvas e retas tangentes
# z = f(x,y) = 16-4x²-y²
# Determinar fx(1,1) e fy(1,1)  P = (1,1) neste ponto z = 11
eixox = np.arange(-3, 3, 0.1)   # x variando de -2 a 2
eixoy = np.arange(-np.sqrt(3), np.sqrt(3), 0.1)   # y variando de -raiz(2) a raiz(2)
x, y = np.meshgrid(eixox,eixoy)  # cria a grade (domínio)
z = 16 -4*(x**2) -y**2  # z é a função: parabolóide

# o plano y = 1 intercepta o parabolóide 
eixoz = np.arange(-10, 10, 0.1)
xplanoy, zplanoy = np.meshgrid(eixox,eixoz)
# resultando na curva na C1: z = 2 - x²
C1 = 15 - 4*(x**2)
# T1 é a reta tangente à curva C1 em P
# a inclinação de T1 é fx = -2x e T1 passa por P = (1,1,11)
T1 = -8*x + 19

# o plano x = 1 intercepta o parabolóide 
eixoz = np.arange(-10, 10, 0.1)
yplanox, zplanox = np.meshgrid(eixoy,eixoz)
# resultando na curva C2: z = 3 - 2y²
C2 = 12 - y**2
# T2 é a reta tangente à curva C2 em P
# a inclinação de T2 é fy = -4x e T2 passa por P = (1,1,11)
T2 = -2*y + 14

# Plano tangente ao parabolóide em P = (1,1,11)
plano = -8*x - 2*y +21

fig = plt.figure(figsize=(6,6))          # cria o objeto figura
ax = Axes3D(fig)            # cria eixos 3d para a figura
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  
ax.plot_wireframe(x, y, z, rstride=1, cstride=1, linewidth=1, color='b') 
ax.plot_surface(xplanoy, 1, zplanoy, rstride=1, cstride=1, cmap=cm.viridis)
ax.plot_wireframe(x, 1, C1, rstride=1, cstride=1, linewidth=5, color='k') 
ax.plot_surface(1, yplanox, zplanox, rstride=1, cstride=1, cmap=cm.viridis)
ax.plot_wireframe(1, y, C2, rstride=1, cstride=1, linewidth=5, color='k')  
ax.plot_wireframe(x, 1, T1, rstride=1, cstride=1, linewidth=5, color='r') 
ax.plot_wireframe(1, y, T2, rstride=1, cstride=1, linewidth=5, color='r') 
ax.plot_wireframe(x, y, plano, rstride=2, cstride=2, linewidth=1, color='g')  

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.set_zlabel('eixo z')
plt.show()
```
