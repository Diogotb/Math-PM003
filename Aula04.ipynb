```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm 
```

```python
# Exemplo
eixox = np.arange(-4, 4, 0.5)  # min, max, passo
eixoy = np.arange(-4, 4, 0.5)  # min, max, passo
x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domÃ­nio)
z = 25 - x**2 - y**2   # z = f(x,y) = 25 - xÂ² - yÂ²

fig1 = plt.figure(figsize=(4,4))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
plt.show()

fx = -2*x
fy = -2*y

# grÃ¡fico do mapa dos vetores gradiente
fig, ax = plt.subplots(figsize=(5,5))
ax.quiver(x,y,fx,fy)  # (gradiente de f)(x,y) = [fx, fy]
ax.set_aspect('equal')

# grÃ¡fico das curvas de nÃ­vel
CS = plt.contour(x,y,z,10) 
plt.clabel(CS, inline=1, fontsize=8)
plt.show()
```

```python
# ExercÃ­cio da Lista 1 (a)
eixox = np.arange(-4, 4, 0.5)
eixoy = np.arange(-4, 4, 0.5)
x, y = np.meshgrid(eixox, eixoy)  # cria a grade (domÃ­nio)
z = (y - 2*x)**2

fig1 = plt.figure(figsize=(4,4))          # cria o objeto figura
ax1 = Axes3D(fig1)                         # cria eixos 3d para a figura
ax1.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.viridis)  # estabelece ordem dos eixos
plt.show()

fx = -4*(y - 2*x)
fy = 2*(y - 2*x)

fig, ax = plt.subplots(figsize=(5,5))
ax.quiver(x,y,fx,fy)
ax.set_aspect('equal')

CS = plt.contour(x,y,z,30) 
plt.clabel(CS, inline=1, fontsize=8)
plt.show()
```

```python
# ExercÃ­cio da Lista 1 (b)
fig, ax = plt.subplots(figsize=(4,4)) 
x = np.arange(-6.0, 6.0, 0.1) 
# k = 0
y01 = x
y02 = -x
ax.plot(x, y01, color = "black")  
ax.plot(x, y02, color = "black")  
cores = ['blue', 'green', 'orange', 'red']
for k in range(1, 5):
    y1 = k*np.sqrt((1 + (x**2)/(k**2)))
    y2 = -k*np.sqrt((1 + (x**2)/(k**2)))   
    ax.plot(x, y1, color = cores[k-1])  
    ax.plot(x, y2, color = cores[k-1])  
ax.grid(True, linestyle='-')  # formato da grade de fundo (opcional)
plt.show()
```

```python
# ExercÃ­cio da Lista 1 (c)
fig, ax = plt.subplots(figsize=(4,4)) 
# coordenadas polares para parametrizaÃ§Ã£o das circunferÃªncias
# xÂ² + yÂ² = raioÂ² = rÂ² - (c/k)Â²
# x = raio*cos(t)
# y = raio*sen(t)
# 0 <= t <= 2*pi
c = 1
r = 4
curvas = [0.5, 1, 2, 100]
cores = ['blue', 'green', 'orange', 'red']
t = np.linspace(0, 2*np.pi, 100)
ax.plot([0], [0], "ko")
for k in range(4):
    raio = r**2 - (c/curvas[k])**2
    x = raio*np.cos(t)
    y = raio*np.sin(t) 
    ax.plot(x, y, color = cores[k])  
ax.grid(True, linestyle='-')  # formato da grade de fundo (opcional)
plt.show()
```
