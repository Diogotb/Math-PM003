```python
%matplotlib inline

# bibliotecas
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

# cria figura e eixos 3D
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

# esfera para o corpo (x, y, z)
rho = 4
# parâmetros (u, v) = (phi, theta)
phi = np.linspace(0, np.pi, 30)
theta = np.linspace(0, 2*np.pi, 30)
phi, theta = np.meshgrid(phi, theta)  # define o domínio D (produto cartesiano)
# parametrização
x = rho*np.sin(phi)*np.cos(theta) 
y = rho*np.sin(phi)*np.sin(theta) 
z = rho*np.cos(phi)
# gráfico
ax.plot_surface(x, y, z+4, alpha=1, color='aliceblue')

# cachecol
rho = 4.1
# parâmetros
phi = np.linspace(0, np.pi/4, 30)
theta = np.linspace(0, 2*np.pi, 30)
phi, theta = np.meshgrid(phi, theta)
# parametrização
x = rho*np.sin(phi)*np.cos(theta) 
y = rho*np.sin(phi)*np.sin(theta) 
z = rho*np.cos(phi)
# gráfico
ax.plot_surface(x, y, z+4.1, alpha=1, color='red')

# esfera para a cabeça
rho = 2
# parâmetros
phi = np.linspace(0, np.pi, 30)
theta = np.linspace(0, 2*np.pi, 30)
phi, theta = np.meshgrid(phi, theta)
# parametrização
x = rho*np.sin(phi)*np.cos(theta) 
y = rho*np.sin(phi)*np.sin(theta) 
z = rho*np.cos(phi)
# gráfico
ax.plot_surface(x, y, z+10, alpha=1, color='aliceblue')

# aba do chapéu
raio = 3
e = 11.5
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, raio, 30)
r, theta = np.meshgrid(r, theta)
# parametrização
x = r*np.cos(theta)
y = r*np.sin(theta)
z = e + 0*r
# gráfico
ax.plot_surface(x, y, z, alpha=1, color='black')

# cilindro para formar a lateral da fita do chapéu 
a = 2.25
b = 11.5
c = 12
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
z = np.linspace(b, c, 30)
z, theta = np.meshgrid(z, theta)
# parametrização
x = a*np.cos(theta)
y = a*np.sin(theta) 
z = z
# gráfico
ax.plot_surface(x, y, z, alpha=0.6, color='red')

# plano para base superior da fita do chapéu
a = 2.25
e = 12
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, a, 30)
r, theta = np.meshgrid(r, theta)
# parametrização
x = r*np.cos(theta)
y = r*np.sin(theta)
z = e + 0*r
# gráfico
ax.plot_surface(x, y, z, alpha=1, color='red')

# plano para base inferior da fita do chapéu
a = 2.25
e = 11.5
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, a, 30)
r, theta = np.meshgrid(r, theta)
# parametrização
x = r*np.cos(theta)
y = r*np.sin(theta)
z = e + 0*r
# gráfico
ax.plot_surface(x, y, z, alpha=1, color='red')

# hiperbolóide para a parte de cima do chapéu
a = 2
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
v = np.linspace(0, 0.5, 30)
theta, v = np.meshgrid(theta, v)
# parametrização
x = a*np.cos(theta)*np.cosh(v)
y = a*np.sin(theta)*np.cosh(v)
z = (a**2)*np.sinh(v)
# gráfico
ax.plot_surface(x, y, z+12, alpha=1, color='black')

# plano para o topo do chapéu
a = 2.25
e = 14
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, a, 30)
r, theta = np.meshgrid(r, theta)
# parametrização
x = r*np.cos(theta)
y = r*np.sin(theta)
z = e + 0*r
# gráfico
ax.plot_surface(x, y, z, alpha=1, color='black')

# nariz
a = 1
b = 0.4  # raio
c = 4  # comprimento
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, a, 30)
theta, r = np.meshgrid(theta, r)
# parametrização
x = b*r*np.cos(theta)
z = b*r*np.sin(theta)
y = c*r
# gráfico
ax.plot_surface(x, y-6, z+9.75, alpha=1, color='orange')

# plano para base do chão
a = -10
b = 10
c = -10
d = 10
e = -0.1
# parâmetros
u = np.linspace(a, b, 30)
v = np.linspace(c, d, 30)
u, v = np.meshgrid(u, v)
# parametrização
x = u
y = v
z = e + 0*u
# gráfico
ax.plot_surface(x, y, z, alpha=1, color='brown')

# árvore menor
a = 1
b = 2  # raio
c = 7  # comprimento
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, a, 30)
theta, r = np.meshgrid(theta, r)
# parametrização
x = b*r*np.cos(theta)
y = b*r*np.sin(theta)
z = -c*r
# gráfico
ax.plot_surface(x+7, y, z+7, alpha=1, color='green')

# árvore maior
a = 1
b = 2   # raio
c = 11  # comprimento
# parâmetros
theta = np.linspace(0, 2*np.pi, 30)
r = np.linspace(0, 1, 30)
theta, r = np.meshgrid(theta, r)
# parametrização
x = b*r*np.cos(theta)
y = b*r*np.sin(theta)
z = -c*r
# gráfico
ax.plot_surface(x-7, y, z+11, alpha=1, color='green')

# título
plt.title("FELIZ NATAL!")
# remove marcações nos eixos x, y, z
ax.set_xticks([])  
ax.set_yticks([])  
ax.set_zticks([])  

# limite dos eixos x e y
plt.xlim([-7, 7])  
plt.ylim([-7, 7])

# Fundo invisível
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')

# ângulo de visualização
ax.view_init(0, 300)

plt.show()
```
